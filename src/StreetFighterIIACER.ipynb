# Configurazione

```
import retro
from stable_baselines.common.policies import MlpPolicy, MlpLstmPolicy, MlpLnLstmPolicy, CnnLnLstmPolicy, CnnPolicy, CnnLstmPolicy
from stable_baselines.common.vec_env import SubprocVecEnv, DummyVecEnv
from stable_baselines.common.schedules import LinearSchedule
from stable_baselines import results_plotter
from stable_baselines import PPO2, A2C, ACER
import numpy as np
import gym
```

```
sts =  [#'RC-Mid',
        'RZ-Mid',
        #'RD-Mid',
        'Car',
        #'RB-Mid',
        #'RH-Mid',
        #'Bricks',
        #'RR-Mid',
        #'RG-Mid',
        #'RK-Mid',
        #'Barrel',
        #'RBal-Mid',
        'RV-Mid',
        #'RS-Mid',
        'RMb-Mid']

gamename = 'StreetFighterIISpecialChampionEdition-Genesis'
```

Per ACER è necessario creare un'ambiente dicreto

```
#Street fighter discretizer
class SFDiscretizer(gym.ActionWrapper):
  
    def __init__(self, env):
        super(SFDiscretizer, self).__init__(env)
        buttons = ["B", "A", "MODE", "START", "UP", "DOWN", "LEFT", "RIGHT", "C", "Y", "X", "Z"]
        actions = ["B", "A", ["UP"], ["DOWN"], ["LEFT"],["LEFT","DOWN"],["RIGHT","DOWN"], ["RIGHT"], "C", "Y", "X", "Z"]
        self._actions = []
        for action in actions:
            #print(action)
            arr = np.array([False] * 12)
            for button in action:
                #print(button)
                arr[buttons.index(button)] = True
            self._actions.append(arr)
        self.action_space = gym.spaces.Discrete(len(self._actions))

    def action(self, a): # pylint: disable=W0221
        return self._actions[a].copy()
```

```
n_cpu = 2

#env.close()
#Champion.Level1.RyuVsGuile
env = SFDiscretizer(retro.make('StreetFighterIISpecialChampionEdition-Genesis', state='RR-Easy' ,scenario='scenario2',record='/gdrive/My Drive/MachineLearning/StreetFighterII/ACER'))
env = SubprocVecEnv([lambda: env])
```

```
model = ACER.load("/gdrive/My Drive/MachineLearning/StreetFighterII/ACER/finalMidStreetfighterIIACER.pkl", env,tensorboard_log="/gdrive/My Drive/MachineLearning/StreetFighterII/ACER/")
#model.set_env(env)
```

L'agente viene addestrato per 2048 step su CnnPolicy e il reso dei parametri di default, come reward function viene usato il punteggio di gioco codificato nello scenario2

```
#ACER model and training single-stage
modelname = 'finalMidStreetfighterIIACER'
model = ACER(CnnPolicy,env,n_steps=2048,verbose=1,tensorboard_log="/gdrive/My Drive/MachineLearning/StreetFighterII/ACER/")

print('Start training for arcade')
#env = SubprocVecEnv([lambda: retro.make('StreetFighterIISpecialChampionEdition-Genesis', state='Champion.3S.Level1.RyuVsGuile', scenario='scenarioLua')])
#env = Monitor(env,'/gdrive/My Drive/MachineLearning/StreetFighterII/PPO')
env = SFDiscretizer(retro.make('StreetFighterIISpecialChampionEdition-Genesis', state='Champion.Level1.RyuVsGuile' ,scenario='scenario2'))
env = SubprocVecEnv([lambda: env])
model.set_env(env)
model.learn(total_timesteps=1000000)
model.save("/gdrive/My Drive/MachineLearning/StreetFighterII/ACER/" + modelname)
env.close()
print("End training")
   
```

Sezione con addestramento multi stage

```
# ACER model and training multi stage
modelname = 'streetfighterIIACER'
model = ACER(CnnPolicy,env,n_steps=512,verbose=1,tensorboard_log="/gdrive/My Drive/MachineLearning/StreetFighterII/ACER/")
#env.close()

for st in sts:
  print("Start training in state:")
  print(st)
  env = SFDiscretizer(retro.make(gamename, state=st,scenario='scenarioLua'))
  env = SubprocVecEnv([lambda: env])
  model.set_env(env)
  model.learn(total_timesteps=100000,tb_log_name=st)
  model.save("/gdrive/My Drive/MachineLearning/StreetFighterII/ACER/" + modelname)
  env.close()
  print("End training episode")
  


#model.learn(total_timesteps=1000000)
model.save("/gdrive/My Drive/MachineLearning/StreetFighterII/ACER/" + modelname)
```
